```python
import numpy as np
import pandas as pd
```

# Loading Data

```python
df = pd.read_csv('balance-scale.csv')
```

```python
df
```

# Data Analysis

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['Class'].unique()
```

```python
class_ids = {"B":0, "R":1, "L":2}
```

```python
df_trimmed = df.drop(['Class'],axis=1)
df_trimmed
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
normalized_data = scaler.fit_transform(df_trimmed)
```

```python
df_nd = pd.DataFrame(normalized_data)
df_nd
```

```python
df['balance_ids'] = df.Class.map(class_ids)
```

```python
df
```

# Testing and Training

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(df[['L-Weight', 'L-Distance', 'R-Weight' , 'R-Distance']], df['balance_ids'], test_size = 0.30 , random_state = 0, shuffle = False)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(random_state = 0)
```

```python
clf.fit(X_train, Y_train)
```

```python
Y_predict = clf.predict(X_test)
```

```python
Y_predict
```

# Evaluation Methods

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
mean_squared_error(Y_test , Y_predict)
```

```python
confusion_matrix(Y_test, Y_predict)
```

```python
accuracy_score(Y_test, Y_predict)
```

# HyperParameter Tuning

```python
param_dist = {
    "criterion":["gini", "entropy"],
    "max_depth":[1,2,3,4,5,6,7,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(clf, param_grid = param_dist, cv = 10, n_jobs = -1)
```

```python
grid.fit(X_train, Y_train)
```

```python
# For Calculating Best Accuracy
grid.best_score_
```
